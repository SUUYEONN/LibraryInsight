```python
# 특정 기간 (2022년 7월 1일부터) 데이터 선택
start_date = pd.to_datetime('2022-07-01')
filtered_df = df[df['날짜'] >= start_date]
filtered_posts_per_day = filtered_df['날짜'].dt.date.value_counts().sort_index()

# 날짜별 게시물 수 시각화
plt.figure(figsize=(12, 6))
plt.plot(filtered_posts_per_day.index, filtered_posts_per_day.values, label='Number of Posts', color='b', linewidth=2)

# 특정 날짜 강조 (2022년 10월 13일)
highlight_date = pd.to_datetime('2022-10-13')
highlight_value = filtered_posts_per_day[highlight_date.date()]

# 마커 추가
plt.scatter([highlight_date], [highlight_value], color='red', zorder=5)

# 주석 추가 (화살표로 강조)
plt.annotate('도서관 개관일: 2022-10-13', xy=(highlight_date, highlight_value), xytext=(highlight_date, highlight_value + 10),
             arrowprops=dict(facecolor='red', shrink=0.05, width=2, headwidth=8), fontsize=12, color='red')

plt.title('Number of Posts per Day (Zoomed View)')
plt.xlabel('Date')
plt.ylabel('Number of Posts')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

### 게시글이 작성된 시간대 분석

```python
# 시간대 추출
df['시간'] = df['날짜'].dt.hour

# 시간대별 게시물 수
posts_per_hour = df['시간'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(posts_per_hour.index, posts_per_hour.values, color='c')
plt.title('Number of Posts per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Posts')
plt.xticks(np.arange(0, 24, 1))
plt.grid(True)
plt.tight_layout()

plt.show()
```

주로 10~20시 사이의 학생들이 학교에 많이 있는 시간에 게시글이 많이 올라온다. 

### 시간대별 평균 좋아요 수

```python
# 시간대별 평균 좋아요 수
average_likes_per_hour = df.groupby('시간')['좋아요'].mean()
average_likes_per_hour
```

```python
plt.figure(figsize=(12, 6))
plt.plot(average_likes_per_hour.index, average_likes_per_hour.values, marker='o', linestyle='-', color='m')
plt.title('Average Likes per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Likes')
plt.xticks(np.arange(0, 24, 1))
plt.grid(True)
plt.tight_layout()

plt.show()
```

### 요일별 게시물 수 분석

```python
# 요일 추출
df['요일'] = df['날짜'].dt.day_name()

# 요일 순서 지정
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# 요일 컬럼을 Categorical로 변환하여 순서 지정
df['요일'] = pd.Categorical(df['요일'], categories=day_order, ordered=True)

# 요일별 게시물 수
posts_per_day_of_week = df['요일'].value_counts().sort_index()
posts_per_day_of_week
```

```python
# 요일별 게시물 수 시각화
plt.figure(figsize=(10, 5))
posts_per_day_of_week.plot(kind='bar')
plt.title('Number of Posts by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Posts')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

학생들이 학교에 있는 요일에 게시글이 많다. 
- 교내 특성상 금요일에 공강이 많기 때문에 금요일부터 도서관과 관련된 게시글을 적어진다. 

### 게시물 길이와 좋아요 수의 상관관계 분석

```python
# 게시물 길이 계산
df.loc[:, '글자수'] = df['내용'].str.len()

# 게시물 길이와 좋아요 수의 상관관계 계산
correlation = df['글자수'].corr(df['좋아요'])
correlation
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['글자수'], df['좋아요'], alpha=0.5)
plt.title('Correlation between Post Length and Likes')
plt.xlabel('Post Length')
plt.ylabel('Likes')
plt.show()
```
